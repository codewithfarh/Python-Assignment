Ques :  6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It 
should print the date in the form March 12, 2021. 

```python
# enter code.
date_input = input("enter date : (mm/dd/year) :")

parts = date_input.split("/")

month_num = int(parts[0])

date = (parts[1])

year = (parts[2])

month = ["january","february","march","april","may","june","july","august","september","october","november","december"]

month_name = month[month_num-1]

print(f"{month_name},{date},{year}")
```

Ques : 8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be 
like this : 
Sum of row 1 = 32 
Sum of row 2 = 31 
Sum of row 3 = 63 

```python
# enter code.

matrix = [
    [10, 10, 12],
    [5, 20, 6],
    [30, 20, 13]
]

for i in range(len(matrix)):
    
    row_sum = sum(matrix[i])
    
    
    print("Sum of row", i + 1, "=", row_sum)
```

Ques : 9. Write a program to add two matrices of size n x m. 

```python
# enter code.
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8, 9],
    [2, 1, 0]
]
result = [
    [0, 0, 0],
    [0, 0, 0]
]

for i in range(len(A)):
    
    for j in range(len(A[0])):
        result[i][j] = A[i][j] + B[i][j]


print("Sum of two matrices:")
for r in result:
    print(r)
```

Ques : 10. Write a program to multiply two matrices.

```python
# enter code.

A = [[1, 2], 
     [3, 4]]

B = [[5, 6], 
     [7, 8]]

result = [[0, 0], 
          [0, 0]]


for i in range(2):      
    for j in range(2):  
        for k in range(2): 
            result[i][j] = result[i][j] + (A[i][k] * B[k][j])


print("Matrix_Multiplication ka result:")
for row in result:
    print(row)
```
