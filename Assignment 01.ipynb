3. Write a program in which use all the operators we can use in Python

```python
# Write your code 
a = 10
b = 6
c = [1,2,3]
d = [1,2,3,4]

print(f"innitial variables a = {a} b = {b} c = {c} d = {d} \n")

print(f"---Arithmatic operators---")
print(f" a + b (addition): {a+b}")
print(f" a - b  (subtration): {a-b}")
print(f" a * b  (multiply): {a*b}")
print(f" a / b (divide): {a/b}") #answer in float values.
print(f" a ** b (power): {a**b}") # a to the power of b.
print(f" a // b (floor division): {a//b}") # answer in round off in int value.
print(f" a % b (modulo): {a%b}") # answer in remainder.
print(f" a > b (greater than): {a>b}") # boolean value. (True).
print(f" a < b (less than): {a<b}") # boolean value. (False).
```

4. Completes the following steps of small task:
    - Mention Marks of English , Islamiat and Maths out of 100 in 3 different variables
    - Mention Variable of Total Marks and assign 300 to it
    - Calculate Percentage

```python
# Write your code here
english = 65
islamiat = 72
mathematics = 77

Total_marks = 300
Obtained_marks = (english + islamiat + mathematics) *100 / 300

print (f"-----results-----")
print (f" english : {english} ")
print (f" islamiat : {islamiat}")
print (f" mathematics : {mathematics} ")
print (f"Total_marks : {Total_marks}")
print (f" Obtained_marks : {Obtained_marks}")

if Obtained_marks >= 70:
    print( "Grade A")
elif Obtained_marks <= 30:
    print("Grade C")
else :
    print ("Grade D")
```
