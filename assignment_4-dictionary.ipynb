Q4 — Count Frequency of Numbers
numbers = [1,2,2,3,3,3,4,4,4,4]
Create frequency dictionary.

```python
# your code.
numbers = [1,2,2,3,3,3,4,4,4,4]

frequency = {}

for n in numbers:
    if n in frequency:
        frequency [n] = frequency [n] + 1

    else :
        frequency [n] = 1

print(frequency)
```

Q5 — Word Counter
Take sentence input and count word frequency.

```python
# your code .
sentence = input("enter a sentence :")

word = sentence.split()

frequency = {}

for i in word:
    if i in frequency:
        frequency[i] = frequency[i] + 1

    else:
        frequency[i] = 1


print (frequency)

```

Q6 — Student Marks & Average
marks = {"Math":80, "English":70, "Physics":90}
Calculate total and average.

```python
# your code.
marks = {
    "math" : 80,
    "english" : 70,
    "physics" : 90
}

total = sum(marks.values())

average = total / len(marks)

print(f"average :  {average} ")
print(f" total : {total}")
```

Q7 — Highest Scoring Student
students = {
"Ali":85,
"Sara":92,
"Ahmed":78,
"Zara":95
}
Find highest scoring student.


```python
# your code .
students = {
    "Ali" : 85,
    "Sara" : 92,
    "Ahmed" : 78,
    "Zara" : 95,
}

highest_score = max(students , key=students.get)

highest_marks = students.get(highest_score)

print(f"the highest score is : {highest_score}")
print(f"the highest marks is :{highest_marks}")
```
